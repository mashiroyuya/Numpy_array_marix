```python
%matplotlib inline
import numpy as np
from numpy import random
from matplotlib import pyplot
from matplotlib import animation
```

```python
nmax = 10 # 粒子の個数
tmax = 100# 最大時間

ll = np.array([10,10]) # 領域

h = 0.001 # 時間刻み幅
```

```python
def force(r):
    n = len(r)
    #力の初期化
    f = np.zeros((n,2))
    
    for i in range(0,n-1):
        ri = r[i]
        fi = f[i]
        for j in range(i+1,n):
            rij = r[j] - r[i] #i番目とj番目の距離
            rij -= np.round(rij/ll)*ll #周期境界条件
            lenr = np.linalg.norm(rij) #rijの大きさを求める
            fij = (1.0/(lenr**3)) * rij #力の大きさ r^2に反比例
            fi += fij #力を加算
            f[j] -= fij # 作用反作用の法則を考慮
        f[i] = fi
    return f
          
```

```python
#初期化
ballr = np.array([[4.,5.],[6.,5.]])#random.rand(nmax,2) * ll
oldr = ballr
v = np.zeros((2,2))
ballf = np.zeros((2,2))
```

```python
f = force(ballr)
```

```python
ballr = oldr + v*h + f*h**2 / 2.0
```

```python
fig = pyplot.figure(figsize=(5,5))
images = []
for t in range(0,tmax):
    k = 0.0
    
    f = force(ballr)
    
    newr = 2.0 * ballr - oldr + f*h**2
    
    v = v + f*h
    for vv in v:
        k += vv.dot(vv)
    k /= 2.0
    
    oldr = ballr
    ballr = newr
    
    ballr -= np.round(ballr/ll)*ll
    
    for b in ballr:
        print(t)
        pyplot.scatter(b[0],b[1])
    pyplot.show()
    #pyplot.scatter(t,k)
```
