## モジュールのインポート
モジュールとは便利な関数パッケージの様なもの(だいぶ適当)．有名なモジュールを下に上げておく．
* **numpy**　数学関係，例えばpiやe，行列計算などの関数や命令が入っている．我々には絶対必要！
* **scipy**　科学関係，最小二乗法やフーリエ変換，球面調和関数などの関係や命令が入っている．やっはり必要！
* **matplotlib**　グラフ関係，論文にそのまま載せれるレベルの図が描ける．もちろんepsやpdf,jpgなどに対応．動画も作成できる
* **sympy**　解析解が出せる(numpyやscipyは数値解)．微分積分や因数分解，階乗計算，順列組み合わせなどほぼなんでもできる
* **pandas**　データマイニングなどの解析に使える．
* **sys**　システム系をいじれる…よく知らない
* **os**　OS依存の機能を呼び出せる
* **urllib**　サーバーからデータ取ってきたりアタックしたり…
* **etc…**

よく使うものを下に記します

```python
print(np.pi)
print(np.e)
print(np.ones(5))
print(np.ones((7,3)))
print(np.zeros((5,3)))
print(np.array([[1.,2.,3.],[4.,5.,6.],[7.,7.,7.]]))
mat1 = np.ones((4,3))
mat1[0,0] = 5
print(mat1)
print(mat1.shape)
print(mat1.shape[0])
print(mat1.dtype)
print(np.linspace(0,10,10)) #最後の数字はデータ総数を指定
print(np.arange(0,10.5,0.5)) #最後の数字はステップ数を指定
mat2 = np.linspace(0,10,11)
mat3 = np.arange(11,22,1.)
mat4 = np.r_[mat2,mat3]
mat5 = np.c_[mat2,mat3]
print(mat4)
print(mat5.shape)
print(np.identity(5))
print(np.eye(5,3))
a, b = np.meshgrid([1,2,3], [4,5,6,7])
print(a)
print(b)
mat6 = np.arange(0,10,0.1)
print(mat6[1])
print(mat6[-1])
print(mat6[4:10])
print(mat6[:5])
print(mat6[40:])
print(np.sum(mat6))
print(np.average(mat6))
print(np.std(mat6))
```

## if文for文while文

```python
s = 10

if s==5 :
    print('a')
elif s==2:
    print('b')
elif s==0:
    print('オワタ')
else:
    print('f')

s = 0
for i in np.arange(11):
    s += i # s = s + 1 の意味
    #print(i)
print(s)

s = 0
while s < 10:
    print(s)
    s += 1
```

## ループ文使ったら負け(遅くなるの意味)

```python
import time
import random

hkl = np.array([0.5,0.5,0.5])
test = np.ones((10000000,3))

for i in range(100) :
    for j in range(3) :
        test[i,j] = random.random()
print (test.shape)

all_data = []

start1 = time.time()
for i in range(test.shape[0]):
    all_data.append(test[i,0]*hkl[0]+test[i,1]*hkl[1]+test[i,2]*hkl[2])
print('-'*10)
print('ループ計算')
print(sum(np.array(all_data)))
print(str(time.time()-start1)+' sec.')
print('-'*10)
print('行列計算')
start2 = time.time()
test2 = sum(np.dot(test, hkl))
print(test2)
print(str(time.time()-start2)+' sec.')
```
